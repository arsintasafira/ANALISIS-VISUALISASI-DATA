Dataset ini mencakup berbagai fitur yang terkait dengan preferensi belanja pelanggan, 
mengumpulkan informasi penting bagi bisnis yang ingin meningkatkan pemahaman mereka tentang basis pelanggan mereka. 
Fitur-fitur tersebut meliputi usia pelanggan, jenis kelamin, jumlah pembelian, metode pembayaran yang disukai, 
frekuensi pembelian, dan peringkat umpan balik. Selain itu, data tentang jenis barang yang dibeli, frekuensi belanja, musim belanja yang disukai, dan interaksi dengan penawaran promosi juga disertakan. 
Dengan 3900 data, dataset ini berfungsi sebagai fondasi bagi bisnis yang ingin menerapkan wawasan berbasis data untuk pengambilan keputusan yang lebih baik dan strategi yang berpusat pada pelanggan.

Daftar Dataset (berdasarkan kolom)
ID Pelanggan - Tanda pengenal unik untuk setiap pelanggan
Usia - Usia pelanggan
Jenis Kelamin - Jenis kelamin pelanggan (Pria/Wanita)
Item yang Dibeli - Item yang dibeli oleh pelanggan
Kategori - Kategori barang yang dibeli
Jumlah Pembelian (USD) - Jumlah pembelian dalam USD
Lokasi - Lokasi tempat pembelian dilakukan
Ukuran - Ukuran barang yang dibeli
Warna - Warna barang yang dibeli
Musim - Musim saat pembelian dilakukan
Peringkat Ulasan - Peringkat yang diberikan oleh pelanggan untuk item yang dibeli
Status Langganan - Menunjukkan apakah pelanggan sudah langganan (Ya/Tidak)
Jenis Pengiriman - Jenis pengiriman yang dipilih oleh pelanggan
Diskon Diterapkan - Menunjukkan apakah diskon diterapkan pada pembelian (Ya/Tidak)
Kode Promo Digunakan - Menunjukkan apakah kode promo digunakan untuk pembelian (Ya/Tidak)
Pembelian Sebelumnya - Jumlah total transaksi yang telah dilakukan oleh pelanggan di toko, tidak termasuk transaksi yang sedang berlangsung
Metode Pembayaran - Metode pembayaran yang paling disukai pelanggan
Frekuensi Pembelian - Frekuensi pelanggan melakukan pembelian (mis., Mingguan, Dua Mingguan, Bulanan)

```python
from sklearn.preprocessing import LabelEncoder
```

```python
var_kategori = ['Gender', 'Item Purchased', 'Category', 'Location', 'Size', 'Color', 'Season', 'Subscription Status', 'Payment Method', 'Shipping Type', 'Discount Applied', 'Promo Code Used', 'Preferred Payment Method', 'Frequency of Purchases']
for feature in var_kategori:
        if feature in combined_data.columns.values:
            combined_data[feature] = LabelEncoder().fit_transform(combined_data[feature])
            combined_data[feature] = combined_data[feature].astype('category')
```

```python
data_cont = combined_data.drop(['Gender', 'Item Purchased', 'Category', 'Location', 'Size', 'Color', 'Season', 'Subscription Status', 'Payment Method', 'Shipping Type', 'Discount Applied', 'Promo Code Used','Payment Method', 'Frequency of Purchases'], axis=1)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14,7))
feature_index = 0
for ii in range(2):
    for jj in range(2):
        ax = sns.boxplot(data_cont.iloc[:, feature_index], ax=axes[ii,jj])
        ax.set(title=data_cont.columns.values[feature_index])      
        feature_index += 1
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {'Category': ['Clothing', 'Accessorise', 'Footwear', 'Outerwear']}
df = pd.DataFrame(data)

f, ax = plt.subplots(1, 2, figsize=(18, 8))

Category_counts = df['Category'].value_counts()
num_categories = len(Category_counts)

explode = [0] * num_categories 
explode[1] = 0.1 

Category_counts.plot.pie(explode=explode, autopct='%1.1f%%', ax=ax[0], shadow=True)
ax[0].set_title('Category')
ax[0].set_ylabel('')

Category_counts.plot.bar(ax=ax[1], color=['pink', 'blue'])
ax[1].set_title('Category (Bar Chart)')
ax[1].set_xlabel('Category')
ax[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
